```python
#스프레스시트 (정정)을 2,3학년 데이터를 unpivot하여 (Data_정정)시트에 업데이트

import pandas as pd
from functions import client

using_spreadsheet_id = '1Mi3NQMV_mlZMcgV74O4ZAAy_Xc6wN_cHD8aVs2OCMoM'
spreadsheet = client.open_by_key(using_spreadsheet_id)
sheet = spreadsheet.worksheet('정정')

second_rng = "A1:AH94"
second_data = sheet.get_values(second_rng)
second_df = pd.DataFrame(second_data[1:], columns = second_data[0])
second_unpivot =pd.melt(second_df, id_vars=['학번','이름'], value_vars=second_df.columns[2:], var_name='교과명', value_name='학점')
second_del_idx = second_unpivot[second_unpivot['학점']==''].index
second_unpivot.drop(second_del_idx, inplace=True)

third_rng = "A101:AP194"
third_data = sheet.get_values(third_rng)
third_df = pd.DataFrame(third_data[1:], columns = third_data[0])
third_unpivot =pd.melt(third_df, id_vars=['학번','이름'], value_vars=third_df.columns[2:], var_name='교과명', value_name='학점')
third_del_idx = third_unpivot[third_unpivot['학점']==''].index
third_unpivot.drop(third_del_idx, inplace=True)


all_df = pd.concat([second_unpivot, third_unpivot]).values.tolist()

setsheet = sheet = spreadsheet.worksheet('Data_정정')
setsheet.update(range_name="A2", values=all_df)
```
